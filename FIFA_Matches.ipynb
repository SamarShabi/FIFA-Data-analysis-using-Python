```python
plt.figure(figsize=[10,4])
plt.bar(list(data['nationality'].value_counts()[0:10].keys()), list(data['nationality'].value_counts()[0:10]), color='pink')
plt.title('Nationality')
plt.show()
```

```python
highest = data[['short_name', 'wage_eur']]
```

```python
highest.head()
```

```python
highest = highest.sort_values(by=['wage_eur'],ascending=False)
```

```python
plt.figure(figsize=[12,4])
plt.bar(list(highest['short_name'][0:10]), list(highest['wage_eur'][0:10]), color='green')
plt.title('Wages By Players')
plt.show()
```
